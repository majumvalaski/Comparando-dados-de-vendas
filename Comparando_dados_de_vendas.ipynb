**Passo a passo da soluçcão do Desafio**

```python
# Passo 0 - Entender o desafio que quero resolver

# Passo 1 - Usar o Python para percorrer todos os arquivos da pasta de vendas do drive (pasta basde de dados)
import os 
import pandas as pd

lista_arquivos = os.listdir("/content/drive/MyDrive/Curso Básico Python/01-15 - Curso de Python - Aula 1/Vendas")
display (lista_arquivos)

```

```python
tabela_total = pd.DataFrame()

# Passo 2 - Importar as bases de dados de vendas
for arquivo in lista_arquivos:
  # Se tem vendas no nome do arquivo, eu vou importar o arquivo 
  if "Vendas" in arquivo:
    # Quando eu coloco entre {} eu digo pro python printar o valor/ nome de cada arquivo
    # Importar o arquivo
    tabela = pd.read_csv(f"/content/drive/MyDrive/Curso Básico Python/01-15 - Curso de Python - Aula 1/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela) #add cada tabela na tabela vazia, fazendo uma tabela total com tudo junto

# Passo 3 - Tratar/ compilar as bases de dados
display(tabela_total)
```

```python
# Passo 4 - Calcular o produto mais vendido (em quantidade)
# Quero agrupar os produtos e somar as quantidades vendidas

tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida", "Preco Unitario"]].sort_values(by="Quantidade Vendida", ascending=False)
display(tabela_produtos)
```

```python
# Passo 5 - Calcular o produto que mais faturou (em faturamento)
#criando uma coluna nova para saber o faturamento de cada produto
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario'] 

tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by="Faturamento", ascending=False)
display(tabela_faturamento)
```

```python
# Passo 6 - Calcular a loja/ cidade que mais vendeu (em faturamento) - criar um gráfico/ dashboard
tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']].sort_values(by="Faturamento", ascending=False)
display(tabela_lojas)

```

```python
#Criando um grafico com as informaçoes
import plotly.express as px
#poderia importar la na primeira celula tambem (mais recomendado ate)

grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y= "Faturamento") #precisa chamar o x assim pois o groupby usado antes, faz com que a coluna vire o novo indice da tabela
grafico.show() #exibe o grafico 
```
